```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x 
```

```python
y
```

```python
#Functional Method
plt.plot(x,y)                     # Use plt.show() when using other IDE.
                                  # No need to use in Jupyter can use if u want
plt.xlabel('x Label')
plt.ylabel('y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)                       #plt.subplot(row,column,selected plot)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Programming

figure = plt.figure()
axes = figure.add_axes([0.1,0.1,0.8,0.8])           
                                            #axes = figure.add_axes([left,bottom,width,height]) it consits of a list containing 
                                            #left, bottom, width, height.
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
#Inserting 2 Plots in one plane,

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_xlabel('Plot1 X Label')
axes1.set_ylabel('Plot1 Y Label')
axes1.set_title('Plot1 Title')

axes2.plot(y,x)
axes2.set_xlabel('Plot2 X Label')
axes2.set_ylabel('Plot2 Y Label')
axes2.set_title('Plot2 Title')
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)       #As an Array we can make use of indexing method to plot different graphs/plot 
                                                   #in different location
axes[0].plot(x,y) 
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('Second plot')
plt.tight_layout()
```

```python
axes                  #It consists of an array according to the rows and cols of the subplots.
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)
for current in axes:
    current.plot(x,y)
```

```python
fig = plt.figure(figsize = (8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label = 'X Squared')
axes.plot(x,x**3,label = 'X Cubed')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')

axes.legend(loc = 0)
```

# Plot Appearance

The axes.plot() method consists of various function in it such as                                                                                     
#### axes.plot(1st operand,2nd operand,color,linewidth,alpha,linestyle,marker,markersize,markerfacecolor,markeredgewidth,markeredgecolor)                                                                                                                                                                                                                                               
where 1st and 2nd are opearands                                                                                      
      color           ->  for selecting desired color.                                                                         
      linewidth       ->  gives a desired width to the line.                                                     
      alpha           ->  To manage the transperancy of the line.  
      linestyle       ->  To show a certain line format.                                        
      marker          ->  To select a type of marker to show the points of our data.                                    
      markerstyle     ->  To select the size  of marker                                                                        
      markerfacecolor ->  To give desired color inside the marker.
      markeredgewidth ->  Gives desired width to the marker edge.                                           
      markeredgecolor ->  Gives desired color to the marker edge.
      

```python
fig = plt.figure(figsize = (8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color = 'REd', lw = 2, alpha = 1,linestyle = '--',marker = 'o',markersize = 10,markerfacecolor = 'Yellow',
         markeredgewidth = 2,markeredgecolor = 'Green')
axes.set_title('Plot of data X vs Y',color = 'Black')
axes.set_xlabel('X Data',color = 'Black')
axes.set_ylabel('Y Data',color = 'Black')
plt.show()
plt.tight_layout()
fig.savefig('X_vs_Y_plot.png',dpi = 100)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color = 'Purple',lw = 2,linestyle = '--')

# To select a particular portion of the entire plot.

axes.set_xlim([0,1])
axes.set_ylim([0,2])
```
